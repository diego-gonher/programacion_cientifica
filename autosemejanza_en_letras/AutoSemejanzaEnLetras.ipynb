Inspirado por el video, el objetivo de este cuaderno de Jupyter es poder vizualizar los patrones líricos presentes en cualquier canción que deseemos, haciendo uso de las matrices de auto-semejanza.

## Obteniendo la letra de nuestra canción
Lo primero que debemos hacer es buscar la letra de la canción que deseemos, y luego simplemente la copiamos y pegamos para guardarla en una variable. Podriamos guardar la letra en un archivo .txt del cual podríamos obtenerla, pero aquí solamente copiamos las letras dentro de un String encerrado por """ """. Esta solución es mucho menos elegante, pero también es mucho mas simple.

```
# solo usamos dos paquetes
import matplotlib.pyplot as plt # para graficar
import numpy as np # por los arreglos

# copiamos y pegamos la letra de la canción, no hay necesidad de modificarla
# por eso utilizamos """ """.
lyrics = """ Disrupt and keep dividing
The government is lying
I'm not gonna be a fallacy of this society
Turn off your TV station
That's not real information
Trying to define the human mind
It's like puppetry
And they've got you by your strings
Playing God for money
A sorry sad charade
You can see right through it

Don't wait for anyone
To say it for you
Say it for yourself
What if God was the beggar in disguise?
Would it make any difference?
Would you open your eyes?

Suspend your disbelief
Are you paying close attention?
If you blink then you might miss
Their tricks and their deception
Distrust and disobey the lies they say
Cause all the world's a stage
And if you choose to read the news
Then you must question everything
And they've got you on your knees
They've got your milk and honey
Then we put them on parade
And celebrate what's coming

Don't wait for anyone
To say it for you
Say it for yourself
What if God was the beggar in disguise?
Would it make any difference?
Would you open your eyes?

No kings, no queens
No justice, no peace
Dig deep, believe
The truth will set you free
No kings, no queens
No justice, no peace
Dig deep, and see
The government is lying

Don't wait for anyone
To say it for you
Say it for yourself
Yeah what if God was the beggar in disguise?
Would it make any difference?
Would you open your eyes?

No kings, no queens
No justice, no peace
Dig deep, believe
The truth will set you free
No kings, no queens
No justice, no peace
Would it make any difference
Would you open your eyes? """
```

Ahora debemos separar la letra de la canción en palabras. Para eso usamos el método split(), el cual separa un String en múltiples Strings que están separadas por espacio blanco. Antes de eso, usamos el método lower() para forzar que todas las letras en el String sean minúsculas (lo que facilitará la comparación que haremos).

```
# ahora creamos una lista con cada una de las letras separadas, pero antes
# convertimos todas las letras a minusculas
words = lyrics.lower().split()

# mostrar las primera cinco palabras 
print(words[:5])
```

## Creando la matriz de auto-semejanza
Podemos crear la matriz de auto-semejanza usando el arreglo con las palabras separadas. Para ello, simplemente usamos dos bucles para comparar cada una de las palabras con las otras, y guardamos el resultado en un arreglo bidimensional.

```
# creamos un arreglo 2D vacio que representara la matriz de auto-semejanza 
self_similarity = np.zeros((len(words), len(words)))

# usamos dos bucles para llenar los valores de esta matriz
for i in np.arange(len(words)):

  for j in np.arange(len(words)):

    # comparar las dos letras y el resultado va en el lugar correspondiente
    self_similarity[i][j] = (words[i] == words[j])
```

## Visualización
Finalmente podemos crear la visualización. Simplemente usamos el método imshow() de matplotlib, y añadimos más detalles.

```
# creamos la figura y los ejes
fig1, ax1 = plt.subplots(figsize=(10,10))

# usamos imshow para mostrar la matriz
plt.imshow(self_similarity, cmap='bone')

# añadimos titulo de la cancion arriba
ax1.set_title("Don't Wait", pad=20, fontdict={'fontsize': 30})

# añadimos nombre del artista y album
ax1.set_xlabel("Neck Deep | The Peace and The Panic", fontdict={'fontsize': 20})

# movemos el texto de abajo un poco
ax1.xaxis.labelpad = 20

# desaparecemos los ticks on ambos ejes
ax1.tick_params(which='both', width=0)
plt.xticks(np.arange(0))
plt.yticks(np.arange(0))

# podemos guardar la imagen con la siguiente linea, aca la dejo comentada
#plt.savefig('dont_wait.jpg', dpi=256/2)

# mostramos la grafica
plt.show()
```
